```python
binary_labels = ['fueltypes', 'aspiration', 'doornumbers', 'enginelocation']

ordinal_labels = ['symboling', 'wheelbase', 'carlength', 'carwidth',
                  'carheight', 'curbweight', 'cylindernumber', 'enginesize',
                  'boreratio', 'stroke', 'compressionratio', 'horsepower',
                  'peakrpm', 'citympg', 'highwaympg']

nominal_labels = ['carbody', 'drivewheels', 'enginetype', 'fuelsystem']
```

```python
le = LabelEncoder()

for i in ordinal_labels:
    data[i] = le.fit_transform(data[i])

for i in binary_labels:
    data[i] = le.fit_transform(data[i])
    
data.head()
```

```python
def drop_column(data,column):
    for i in column:
        data.drop(i,axis=1, inplace=True)
```
